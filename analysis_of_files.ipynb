```python
# 5 Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing = (file.isnull().sum() / len(file)) * 100
print(missing)
```

```python
# 6 Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

programming_language = file['Мова програмування']
file = file.dropna(axis=1, how='any')
file.loc[:, 'Мова програмування'] = programming_language

file
```

```python
# 7 Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
missing = (file.isnull().sum() / len(file)) * 100
print(missing)
```

```python
# 8 Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
file = file.dropna()
```

```python
# 9 Визначте новий розмір таблиці за допомогою методу shape
file.shape
```

```python

# 10 Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = file.loc[file['Мова програмування'] == 'Python']
python_data
```

```python
# 11 Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape

```

```python
# 12 Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
grouped = python_data.groupby('Посада')
```

```python
# 13 Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
aggregated = grouped['Зарплата на місяць'].agg(['min', 'max']).reset_index()
aggregated
```

```python
# 14 Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(data):
    return data.mean()

aggregated['avg'] = grouped['Зарплата на місяць'].apply(fill_avg_salary).reset_index()['Зарплата на місяць']
aggregated
```

```python
# 15 Створіть описову статистику за допомогою методу describe для нового стовпчика.
stats = aggregated['avg'].describe()
stats
```

```python
# 16 Збережіть отриману таблицю в CSV файл
aggregated.to_csv("new_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(python_data['Зарплата на місяць'], bins=20, kde=True)
plt.title('Розподіл зарплати на місяць')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Посада', y='Зарплата на місяць', data=python_data, palette='muted')
plt.title('Групування за посадою та розподіл зарплати')
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
min_salary_by_position = aggregated.groupby('Посада')['min'].sum()
plt.pie(min_salary_by_position, labels=min_salary_by_position.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Розподіл мінімальної зарплати за посадами')
plt.show()
```

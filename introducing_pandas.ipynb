```python
# <table class="wikitable collapsible collapsed" style="text-align: right" id="collapsibleTable4">
# 1) Вивести перші рядки таблиці за допомогою методу head
import pandas as pd

match='Коефіцієнт народжуваності в регіонах України'
table = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8',thousands="", decimal=",", match = match,)
t = table[0]
t.head()
```

```python
# 2) Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
t.shape
```

```python
# 3) Замініть у таблиці значення "—" на значення NaN
import numpy as np
# t.replace("—", pd.NA, inplace=True)
t.replace("—", np.nan, inplace=True) 
t.head()
```

```python
# 4) Визначте типи всіх стовпців за допомогою dataframe.dtypes
t.dtypes
```

```python
# 5) Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
# t.fillna(0, inplace=True)
non_numeric_columns = t.columns.difference(['Регіон'])  # Крім колонки "Регіон"
t[non_numeric_columns] = t[non_numeric_columns].apply(pd.to_numeric, errors='coerce')
t.dtypes
```

```python
# 6) Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing = (t.isnull().sum() / len(t)) * 100
print(missing)
```

```python
# 7) Видаліть з таблиці дані по всій країні, останній рядок таблиці
t.drop(len(t)-1, inplace=True)
t.head(28)
```

```python
# 8) Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
withoutreg = t.columns.difference(['Регіон'])  # Крім колонки "Регіон"


t.fillna(t[withoutreg].mean(), inplace=True)
t.head(28)
```

```python
# 9) Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні - 8,1
higher = t[t['2019'] > 8.1]['Регіон'].tolist()
print(higher)
```

```python
# 10) У якому регіоні була найвища народжуваність у 2014 році?
max_2014 = t.loc[t['2014'].idxmax(), 'Регіон']
print(max_2014)
```

```python
# 11) Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році 
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))
plt.bar(t['Регіон'], t['2019'], color='skyblue')
plt.title('Народжуваність по регіонах у 2019 році')
plt.xlabel('Регіон')
plt.ylabel('Коефіцієнт народжуваності')
plt.xticks(rotation=90)
plt.show()
```
